1. What is the difference between a function and a method in Python ?
-  In Python, the terms function and method are often used interchangeably but they have distinct meanings based on context:

Function:

  - A function is a block of reusable code that performs a specific task.
  - Functions can be defined using the def keyword outside of any class.
  - Functions can be called independently, without requiring an instance of a class.
For Example, this is how you define and call a function:


```python
def add(a, b):
    return a + b
    result = add(5, 3)
```

Method:

  - A method is a function that is associated with an object (usually an instance of a class) and is called on that object.
  - Methods are defined within a class and are typically called on instances (or objects) of that class.
  - A method always takes at least one argument, commonly referred to as self, which refers to the instance the method is called on.
  For example, this how you define and call a method:
  

```python
class Calculator:
    def add(self, a, b):
        return a + b
        calc = Calculator()
        result = calc.add(5, 3)
```

2. Explain the concept of function arguments and parameters in Python.
-  In Python, arguments and parameters are essential concepts related to functions. They are used to pass data into functions, allowing the function to operate on different values each time it is called. Let's break down the concepts:

 Parameters:
    - Parameters are the names used in the function definition to represent the values that will be passed to the function when it is called.
    - They act as placeholders for the actual data (arguments) that you provide when calling the function.
For Example:

def greet(name):

    print(f"Hello, {name}!")


In this example, name is a parameter of the greet function. It doesn't hold a specific value until the function is called.

  Arguments:
   - Arguments are the actual values or data you pass to the function when calling it.
   - These values are assigned to the corresponding parameters in the function.
For Example:

greet("Alice")

In this case, "Alice" is the argument passed to the greet function. It is assigned to the name parameter.

3. What are the different ways to define and call a function in Python ?
-   In Python, functions can be defined and called in several ways. Here are the most common methods:

  - Basic Function Definition and Call

You define a function using the def keyword followed by the function name and parentheses. You call the function by using its name and passing the required arguments.


```python
# Function Definition
def greet(name):
    print(f"Hello, {name}!")

# Function Call
greet("Alice")
```

  - Function with Return Value

Functions can return values using the return keyword.


```python
# Function Definition
def add(a, b):
    return a + b

# Function Call
result = add(3, 4)
print(result)
```

  - Lambda Functionn

You can define a function without a name using lambda. These are often used for simple operations that can be written in a single line.


```python
# Lambda Function Definition
square = lambda x: x * x

# Function Call
print(square(5))
```

  - Function as Arguments

Functions can also be passed as arguments to other functions.


```python
# Function Definition
def apply_function(func, value):
    return func(value)

# Lambda function passed as argument
result = apply_function(lambda x: x * x, 5)
print(result)
```

  - Nested Functions

A function can be defined inside another function. The inner function can access variables from the outer function.


```python
# Outer Function
def outer():
    x = 10

    # Inner Function
    def inner():
        print(f"Value of x is {x}")

    inner()

outer()
```

  - Recursive Functions

A function can call itself. This is known as recursion. However, Python has a default recursion limit, so be careful with deep recursion.


```python
# Recursive Function to calculate factorial
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

# Function Call
print(factorial(5))
```

  - Method Functions

You can define methods inside classes, which are functions bound to objects.


```python
# Class Definition
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello, {self.name}!")

# Create an instance of the class
person = Person("Alice")
person.greet()
```

5.  What are iterators in Python and how do they differ from iterables ?
-   In Python, both iterators and iterables are fundamental concepts that deal with the process of iterating (looping) over elements. Here's a breakdown of both:

- Iterable - An iterable is any Python object that can return an iterator. In other words, an iterable is an object you can loop over, such as a list, tuple, or string. It must implement two things:

    - __iter__() method: This method returns an iterator, which is an object that defines the iteration behavior.
    - __getitem__() method (for sequence types like lists, tuples): This allows indexing to access elements.

Common examples of iterables include:

  - Lists ([1, 2, 3])
  - Tuples ((1, 2, 3))
  - Strings ("hello")
  - Dictionaries ({'a': 1, 'b': 2})
  

```python
# Example of an iterable
my_list = [1, 2, 3]
for item in my_list:
    print(item)
```

- Iterator  - An iterator is an object that represents a stream of data. It keeps track of its current position while iterating through the iterable. An iterator implements two methods:

  - __iter__(): This returns the iterator object itself (so the iterator can be used in a for loop).
  - __next__(): This returns the next item from the iterable. When there are no more items left, it raises a StopIteration exception.

Essentially, an iterator is created from an iterable. Once you get an iterator from an iterable, you can use next() to manually retrieve the next element.


```python
# Example of an iterator
my_list = [1, 2, 3]
iterator = iter(my_list)  # Get an iterator from the iterable

print(next(iterator))  # Outputs 1
print(next(iterator))  # Outputs 2
print(next(iterator))  # Outputs 3
# next(iterator) would raise StopIteration since there are no more items
```

6.  Explain the concept of generators in Python and how they are defined.
-   In Python, a generator is a type of iterable, like a list or a tuple, but instead of storing all its items in memory at once, a generator generates the items on the fly as you iterate over it. This is done using the yield keyword, making generators more memory-efficient for working with large datasets or streams of data.

Defining a Generator:

A generator is defined using a function, but instead of using return, you use yield. Every time the generator function is called, it returns an item and pauses, saving the state of the function. The next time you call the generator, it resumes from where it left off. For Example:


```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # Yielding the current value and pausing the function
        count += 1
```

7.  What are the advantages of using generators over regular functions ?
-   Generators offer several advantages over regular functions, especially in scenarios where you need to handle large data sets or perform computations on the fly. Here are the main benefits:

  - Memory Efficiency: Generators are more memory-efficient.
  - Performance: Generators can improve performance in certain cases.
  - Cleaner Code: They simplify iteration and state handling.
  - Infinite Sequences: Ideal for generating infinite sequences.

  For Example:
  

```python
#Regular fuction (returning a list)
def generate_numbers(n):
    return [i for i in range(n)]

# Using it
numbers = generate_numbers(10000)  # All numbers are stored in memory at once
```

```python
#Generator
def generate_numbers(n):
    for i in range(n):
        yield i

# Using it
gen = generate_numbers(10000)
for number in gen:  # Numbers are generated one by one, on demand
    print(number)
```

8.  What is lambda function in Python and when is it typically used ?
-   A lambda function in Python is a small, anonymous function that is defined using the lambda keyword instead of the def keyword. It can take any number of arguments, but it can only have a single expression. The result of that expression is automatically returned by the function.

Lambda functions are most often used in situations where a simple function is required for a short period of time, and you donâ€™t want to define a full function using def. Some common scenarios include:

  - As an argument to higher-order functions: Many functions, such as map(), filter(), and sorted(), accept functions as arguments. You can pass lambda functions as arguments instead of defining a separate named function.

```python
data = [(1, 'apple'), (2, 'banana'), (3, 'cherry')]
sorted_data = sorted(data, key=lambda x: x[1])  # Sort by second element (fruit name)
print(sorted_data)
```

10. What is the difference between 'map()', 'reduce()', and 'filter()' functions in Python  ?
- In Python, map(), reduce(), and filter() are functional programming tools that allow you to operate on iterables (like lists or tuples) in different ways. Here's a breakdown of the differences:

  - map():
    - Purpose: It applies a function to every item of an iterable (e.g., a list) and returns a map object (an iterator) that contains the results.
    - Usage: It's used when you want to transform each element of an iterable individually using a function. For example:
    

```python
numbers = [1, 2, 3, 4]
squared = map(lambda x: x**2, numbers)
print(list(squared))
```

  - filter()
    - Purpose: It applies a function to each item of an iterable and returns an iterator that contains only the items for which the function returns True.
    - Usage: It's used when you want to filter out elements of an iterable based on a condition. For Example:
    

```python
numbers = [1, 2, 3, 4, 5]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))
```

  - reduce()
    - Purpose: It applies a function cumulatively to the items of an iterable, reducing the iterable to a single value. It is available in the functools module.
    - Usage: It's used when you want to "reduce" an iterable into a single value (e.g., sum, product). For Example:
    

```python
from functools import reduce
numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

```python
#5  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, n):
        self.n = n  # Number of terms in the sequence
        self.a, self.b = 0, 1  # Starting values for the Fibonacci sequence
        self.count = 0  # Counter to track how many terms we've yielded

    def __iter__(self):
        return self  # Returns the iterator itself

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration  # Stop iteration if we've yielded the specified number of terms
        # Generate the next term in the Fibonacci sequence
        fib = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return fib
# Create an iterator for the first 6 Fibonacci numbers
fib_iter = FibonacciIterator(6)

# Iterate through the Fibonacci sequence
for num in fib_iter:
    print(num)
```

```python
#6  Write a generator function in Python that yields the power of 2 up to a given exponent.

def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
for power in powers_of_two(3):
    print(power)
```

```python
#9  Write a Python program that uses 'map()' to convert a list of temperatures from Celcius to Fahrenheit.

celsius_temperatures = [0, 20, 30, 40, 100]

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print(fahrenheit_temperatures)
```

```python
#10 Create a Python program that uses 'filter()' to remove all the vowels from a given string.

def is_not_vowel(char):
    return char.lower() not in 'aeiou'

input_string = "This is an example of string without vowels."

filtered_string = ''.join(filter(is_not_vowel, input_string))

print(filtered_string)
```
